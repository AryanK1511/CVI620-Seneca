# OpenCV Basics

Computer vision is a field of artificial intelligence (AI) that uses machine learning and neural networks to teach computers and systems to derive meaningful information from digital images, videos and other visual inputs

```python
# Importing the openCV library
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading an image
img = cv2.imread('photos/test_image.jpg')

# Reading an image in grayscale (flag is 1 by default which means colored image)
img_grayscale = cv2.imread('photos/test_image.jpg', 0)
```

```python
# Print the image data (pixel values), a 2D numpy array
# Each pixel value is 8 bits [0, 255]
print("Original Image: ", img[1])
print("Grayscale Image: ", img_grayscale[1])
```

```python
print(img.shape, img.dtype)
print(img_grayscale.shape, img_grayscale.dtype)
```

The difference you're seeing between the shapes of a colored image (RGB) and a grayscale image is due to the number of channels (dimensions) in each image type.

### RGB Image:
- **Shape**: `(height, width, 3)`
- An RGB (or color) image has three channels: Red, Green, and Blue (hence the 3 in the shape).
- Each pixel in an RGB image is represented by a triplet of values corresponding to the intensity of red, green, and blue for that pixel.

### Grayscale Image:
- **Shape**: `(height, width)`
- A grayscale image has only one channel (no color), so each pixel is represented by a single value that indicates its brightness/intensity.

```python
plt.imshow(img) # To display an image # Uses BGR instead of RGB so matplotlib is inaccurate with colors
# Matplot lib uses colormaps to display data
```

```python
reversed_image = img[:, :, ::-1] # Convert to RGB to get accurate colors
plt.imshow(reversed_image)
```

```python
plt.imshow(img_grayscale) # Without colormap
```
